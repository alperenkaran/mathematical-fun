# How to draw a Sierpinksi Triangle?

1. Fix the `corners` of the triangle.
2. Initialize an empty list called `points`.
3. Initialize the point `p` at a corner.
4. For a random corner `c`, let `p` be the halfway between `p` and `c`.
5. Append `p` to `points`.
6. Repeat step 4.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sierpinski_triangle(n_points, corners=None):
    if corners is None:
        corners = [[0,0], [2,0], [1,np.sqrt(3)]] # the corners of the triangle
        
    points = []
    
    p = corners[np.random.randint(3)]
    
    for i in range(n_points):
        c = corners[np.random.randint(3)]
        p = [(c[0]+p[0])/2, (c[1]+p[1])/2]
        points.append(p)
        
    return np.array(points)
```

```python
points = sierpinski_triangle(5000)

plt.figure(figsize=(6, 6))
plt.scatter(points[:,0], points[:,1], s=1, c='black')
plt.axis('equal')
plt.axis('off')
plt.show()
```

```python
points = sierpinski_triangle(5000, corners=[[0,0],[0,1],[1,0]])

plt.figure(figsize=(6, 6))
plt.scatter(points[:,0], points[:,1], s=1, c='black')
plt.axis('equal')
plt.axis('off')
plt.show()
```
